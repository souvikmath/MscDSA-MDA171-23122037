#create a class pandas Name that help you to read a csv file and print the top /bottom  n rows in the file.

class pandas name
        init()
              stores the data 
        read CSV()
              reads a CSV file
        head()
              prints the top n rows
        tail()
             to print the bottom n rows
                                

```python
class Pandas_Souvik:

    def __init__(self):
        self.file_name =""
        self.file_content =""

    def read_csv(self,file_name):
        self.file_name =file_name
        try:
            with open (file_name,"r+") as file:
                self.file_content =file.readlines()
        except:
            print("file not found")

    def head(self,num_of_rows):
        for i in range (0,num_of_rows):
             print(self.file_content[i].strip())

    def tail( self,num_of_rows):
        start = len(self.file_content)-(num_of_rows+1)
        end =len(self.file_content)-1
        for i in range(start,end):  
            print(self.file_content[i].strip())       

    

        
        
```

```python
pd =Pandas_Souvik()
pd.read_csv("DATA (1).csv")
pd.head(2)
pd.tail(2)
```
